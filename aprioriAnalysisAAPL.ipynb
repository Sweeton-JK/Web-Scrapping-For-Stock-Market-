```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
df = pd.read_csv('./AAPL_stock_data.csv')
```

```python
num_bins = 2
```

```python
df['open_bin'] = pd.cut(df['1. open'], bins=num_bins, labels=False)

df['high_bin'] = pd.cut(df['2. high'], bins=num_bins, labels=False)

df['low_bin'] = pd.cut(df['3. low'], bins=num_bins, labels=False)

df['close_bin'] = pd.cut(df['4. close'], bins=num_bins, labels=False)

df['volume_bin'] = pd.cut(df['5. volume'], bins=num_bins, labels=False)
```

```python
print(df[['open_bin', 'high_bin', 'low_bin', 'close_bin', 'volume_bin']])
```

```python
transactions = df[['open_bin', 'high_bin', 'low_bin', 'close_bin', 'volume_bin']]
```

```python
frequent_itemsets = apriori(transactions, min_support=0.1, use_colnames=False)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```
